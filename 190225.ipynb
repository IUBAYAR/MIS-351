```python
#conditional statements: if/elif/else
nr=352
#let's test nr for divisibility
#division by zero is not defined
print(0,"not defined")
#everything is divisible by 1
print(1, "of course")
if nr % 2 == 0:
  print(2,"divisible!")
elif nr % 3 ==0:
  print(3,"divisible!")
elif nr % 4 ==0:
  print(4,"divisible!")
elif nr % 5 ==0:
  print(5,"divisible!")
elif nr % 6 ==0:
  print(6,"divisible!")
elif nr % 7 ==0:
  print(7,"divisible!")
elif nr % 8 ==0:
  print(8,"divisible!")
else:
   print(9,"divisible!")
```

```python
#in a chain of conditional statements the rest of the code will be omitted
#after a condiition satisfied at some level
```

```python
#while loops runs until the condition is not satisfied anym
```

```python
#it is customary to use counters with while loops
counter = 0
while counter<=9:
  print(counter)
  counter +=1
  #counter = counter +1
```

```python
#for loops
#let's now test divisibility using for loops
nr=403
for i in range(1, 9+1):
  if nr % i == 0:
    print(nr, "is divisible by", i)
  else:
    print(nr, "is not divisible",i)
```

```python
counter = 1
while counter <=9:
  if nr % counter ==0:
    print(nr, "is divisible by", counter)
  else:
    print(nr, "is not divisible by", counter)
  counter +=1
```

```python
#finding prime numbers less than 25
#no number is divisible by something greater than itself
#smallest prime number is
prime=[2]
for i in range(3, 25+1):
  divisors = []
  for j in range(2,i):
    if i%j ==0:
      divisors.append(j)
  if len(divisors)==0:
      prime.append(i)
prime
```
