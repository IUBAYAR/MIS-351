Q1: 200 people were asked about which athletics event they thought was the most exciting to watch. The results are shown as:

Male/100m: 56

*   Male/100m: 56
*   Male/Long jump: 30
*   Male/Javelin: 24
*   Female/100m: 32
*   Female/Long jump: 29
*   Female/Javelin: 29

A person is chosen at random. Given that that person choose 100m, what is the probability that the person was female?

```python
#Importing numpy module
import numpy as np
```

```python
#the modelling of the random variables
#having 100m implies the sample size is 56+32=88
#within these 88 people 32 were female
```

```python
x = np.random.choice(
    ['Male', 'Female'],
    p = [(56)/(56+32), (32)/(56+32)]
)
x
```

```python
#A Monte Carlo Simulation with 1000 Runs
```

```python
outcomes = []
for i in range (10**3):
  outcomes.append(
      np.random.choice(
      ['Male', 'Female'],
      p = [(56)/(56+32), (32)/(56+32)]
      )
  )
```

Q2: There are 18 girls and 12 boys in a class. 2/9 of the girls and 1/4 of the boys walk to school.

* P(orange)=0.25
