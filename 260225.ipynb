Q1: 200 people were asked about which athletics event they thought was the most exciting to watch. The results are shown as:

*   Male/100m: 56
*   Male/Long jump: 30
*   Male/Javelin: 24
*   Female/100m: 32
*   Female/Long jump: 29
*   Female/Javelin: 29

A person is chosen at random. Given that that person choose 100m, what is the probability that the person was female?

```python
#Importing numpy module
import numpy as np
```

```python
#the modelling of the random variables
#having 100m implies the sample size is 56+32=88
#within these 88 people 32 were female
```

```python
x = np.random.choice(
    ['Male','Female'],
    p=[(56)/(56+32), (32)/(56+32)]
)
x
```

```python
#A Monte Carlo Simulation with 1000 runs
```

```python
outcomes = []
for i in range (10**3):
  outcomes.append(
      np.random.choice(
      ['Male','Female'],
      p=[(56)/(56+32), (32)/(56+32)]
      )
  )
```

Q2: There are 18 girls and 12 boys in a class. 2/9 of the girls and 1/4 of the boys walk to school.
One of the students who walks to school is chosen at random. Find the probability that the student is a boy.

```python
sample_space=["boy_walk","girl_walk","boy_not","girl_not"]
18*2/9,12*1/4
```

```python
import pandas as pd
liste=[]

for i in range (10**4):
  x= np.random.choice(
      sample_space,
      p=[3/30, 4/30, 9/30, 14/30]
  )
  liste.append(x)
```

```python
pd.Series(liste).value_counts()
```

```python
pd.Series(liste).value_counts()[3] / pd.Series(liste).value_counts()[2:].sum()
```

Q3: I have a big tub of jelly beans. The probability of picking each different color of jelly bean is shown below:


*   P(red)= 0.2
*   P(yellow)= 0.15
*   P(green)= 0.1
*   P(purple)= 0.3
*   P(orange)= 0.25
  
  If I were to pick 60 jelly beans from the tub, how many orange jelly beans would I expect to pick?

  This time we have a slightly more complex pattern for describing our random variable: the random variable now is the number of orange jelly beans in 60 draws.






```python
sample_space= ["red", "yellow", "green", "purple", "orange"]
#definition of the sub random variable
x=np.random.choice(
    sample_space,
    p=[0.2, 0.15, 0.1, 0.3, 0.25],
    size = 60
)
```

```python
liste = []

for i in range (10**3):
  liste.append(
      np.random.choice(
                      sample_space,
                      p=[0.2, 0.15, 0.1, 0.3, 0.25],
                      size = 60
      )
  )
```
