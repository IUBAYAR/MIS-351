Simpson's Collection

Q1: In packages containing a single random item Simpsons figures are sold for $5. Assuming each figure is equally likely, what is the probability of completing a collection of


*   Homer Simpson,
*   Marge Simpson,
*   Bart Simpson,
*   Lisa Simpson,
*   Maggie Simpson
*   Ned Flanders,
*   Milhouse Van Houten,
*   Montgomery Burns

before spending $100?


```python
sample_space = ["Homer Simpson", "Marge Simpson", "Bart Simpson",
                "Lisa Simpson", "Maggie Simpson", "Ned Flanders",
                "Milhouse Van Houten", "Montgomery Burns"]
probabilities = [1/len(sample_space) for i in range(len(sample_space))]
#a list comprehension with 8 values 1/8
```

```python
liste = []
for i in range(10**3):
  collection = []
  spenditure = 0

  while len(collection) < len (sample_space):
    spenditure +=5
    collection.append (np.random.choice(
                                        sample_space,
                                        p = probabilities
                                        )
                      )
#If you get duplicates you throw them away...
    collection = list (set(collection))

  liste.append(spenditure)
```

```python
liste
```

```python
len ( pd. Series(liste)[pd.Series(liste)< 100]) / len(liste)
```
